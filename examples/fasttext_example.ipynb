````python
import os
import requests
import shutil
import gzip
````

# Load Fasttext Embeddings 

Fasttext has Pre-trained word vectors on English webcrawl and Wikipedia which you can find [here](https://fasttext.cc/docs/en/english-vectors.html) as well as Pre-trained models for 157 different languages which you can find [here](https://fasttext.cc/docs/en/crawl-vectors.html)

````python
filename: str = 'cc.en.300.bin'
gz_filename: str = 'cc.en.300.bin.gz'

#   URL to the FastText embeddings in the language of your choice
url = f'https://dl.fbaipublicfiles.com/fasttext/vectors-crawl/{gz_filename}'

gz_path: str = datapath(gz_filename)
unzipped_path: str = datapath(filename)
````

Downloads the compressed zip of the FastTextEmbeddings

````python
if not os.path.exists(gz_path):
    with requests.get(url, stream=True) as response:
        response.raise_for_status()
        with open(gz_path, 'wb') as file:
            for chunk in response.iter_content(chunk_size=8192):
                file.write(chunk)
````

Unzips the downloaded compressed zip of the FastTextEmbeddings

````python
if not os.path.exists(unzipped_path):
    with gzip.open(gz_path, 'rb') as zipped_file:
        with open(unzipped_path, 'wb') as unzipped_file:
            shutil.copyfileobj(zipped_file, unzipped_file)
````

````python
model = load_facebook_model(unzipped_path)

t = Word2vec(model=model)
output = t.augment('The stories are good', top_n=10)
print(output)
````
