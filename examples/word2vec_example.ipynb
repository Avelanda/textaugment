## Load Google Word2vec embeddings

````python
# Download Google Word2vec embeddings
url: str = 'https://drive.google.com/uc?id=0B7XkCwpI5KDYNlNUTTlSS21pQmM'
gz_filename: str = 'GoogleNews-vectors-negative300.bin.gz'
gz_filepath: str = datapath(gz_filename)

if not os.path.exists(gz_filepath):
    gdown.download(url, gz_filepath, quiet=False)
````

````python
model = KeyedVectors.load_word2vec_format(gz_filepath, binary=True)
````

````python
t = Word2vec(model=model)
output = t.augment('The stories are good', top_n=10)
print(output)
````
