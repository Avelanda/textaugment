```python
from textaugment.transformer.contextual_word_substitution import ContextualWordSubstituion 
from textaugment.transformer.controlled_text_generation import ControlTextGeneration 
from textaugment.transformer.paraphrase import Paraphraser 
from textaugment.transformer.style_transfer_domain_adaption import StyleTransferDomainAdapter 
```

```python
cws = ContextualWordSubstituion()

cws.augment('the quick brown fox jumped over the lazy dog.')
```

```python
ctg = ControlTextGeneration()

ctg.augment('the quick brown fox jumped over the lazy dog.')
```

```python
paraphraser = Paraphraser()

paraphraser.augment('the quick brown fox jumped over the lazy dog.')
```

```python
stda = StyleTransferDomainAdapter()

stda.augment('the quick brown fox jumped over the lazy dog.')
```
